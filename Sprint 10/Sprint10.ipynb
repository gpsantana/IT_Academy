Crea una columna nova que sigui la data de naixament en format Datetime a partir de les columnes dia, mes i any. Utilitzant aquesta columna crea una funció que donada una data, et calculi l'edat actual a dia d'avui.

Utilitza la funció que acabes de crear per generar una columna nova al DataFrame amb l'edat actual.

```python
from datetime import datetime, date

df["fecha"] = (
    df["Día"].astype(str) + "/" +
    df["Mes"].astype(str) + "/" +
    df["Any"].astype(str)
)

df["fecha"] = pd.to_datetime(df["fecha"], format="%d/%m/%Y")

def edad(fnacimiento):
    hoy = date.today()
    return hoy.year - fnacimiento.year - (
        (hoy.month, hoy.day) < (fnacimiento.month, fnacimiento.day)
    )

df["Edad"] = df["fecha"].apply(lambda x: edad(x.date()))

df
```

Utilitzant el següent DataFrame, adjunta la columna "Increment" al dataframe del nivell anterior.

Actualitza la columna salari en funció dels percentatges que s'adjunten. No modifiquis manualment els increments, escriu codi Python per fer les conversions necessàries.

```python
df_increment = pd.DataFrame({
    "Grup Professional": ["Grup A", "Grup B", "Grup C", "Grup D"],
    "Increment": ["5%", "3,5%", "2%", "8%"]
})

df3 = pd.merge(df, df_increment, on="Grup Professional")

df3["Increment"] = (
    df3["Increment"]
    .str.replace("%", "", regex=False)
    .str.replace(",", ".", regex=False)
    .astype(float) / 100
)

df3["Salari mensual"] = df3["Salari mensual"] * (1 + df3["Increment"])

df3
```

Utilitzant un bucle, exporta en 4 fitxers (format .xlsx o .csv) les dades de cada Grup Professional.

```python
grupos = df3["Grup Professional"].unique()

for conjunto in grupos:
    df_grupo = df3.loc[df3["Grup Professional"] == conjunto]
    nombre_grupo = conjunto.replace(" ", "")
    df_grupo.to_excel(f"dades_{nombre_grupo}.xlsx", index=False)

df4 = df3.groupby("Grup Professional").agg({
    "Salari mensual": "mean",
    "Edad": "median",
    "Grup Professional": "count"
}).rename(columns={"Grup Professional": "Count"}).round(2)

df4.to_excel("tabla_resumen.xlsx", index=True)
```

Crea una funció que prengui un dataframe com a paràmetre d'entrada.

La funció ha de crear (i exportar) un gràfic automàticament per a cada columna del dataframe. Per exemple:

un histograma/boxplot si la variable és numèrica
unes barres dels valors més freqüents si és categòrica
unes barres dels anys més freqüents si la dada està en format data.

```python
import seaborn as sns
import os
import matplotlib.pyplot as plt
df=sns.load_dataset("penguins")

def graficos(df, carpeta="graficos_columnas", top_n=10):
    if not os.path.exists(carpeta):
        os.makedirs(carpeta)

    for col in df.columns:
        serie=df[col].dropna()
        dtype=serie.dtype

        if pd.api.types.is_numeric_dtype(dtype):
            plt.figure()
            serie.plot(kind="hist")
            plt.title(col)

        elif pd.api.types.is_object_dtype(dtype):
            conteo=serie.value_counts().head(top_n)
            plt.figure()
            conteo.plot(kind="bar")
            plt.title(col)

        else: 
            pd.api.types.is_datetime64_any_dtype(dtype)
            año=serie.dt.year
            conteo=año.value_counts().sort_values(ascending=False).head(top_n)
            conteo.plot(kind="bar")
            plt.title(col)

        ruta= os.path.join(carpeta, f"{col}.png")
        plt.savefig(ruta)


graficos(df)
```

Fes una funció que donada la matriu de distàncies i la ciutat d'origen, faci una proposta de ruta que sigui el més curta possible que puguis, retornant una llista amb l'ordre de visita. Dóna també la distància total recorreguda.

```python
df5=pd.read_excel(r"C:\Users\GPSantana\Downloads\matriu_distancies.xlsx",header=0,index_col=0)

df5=df5.drop(columns=["Palma","Las Palmas de Gran Canaria"])
df5=df5.drop(index=["Palma","Las Palmas de Gran Canaria"])

df5=df5.fillna(0)

display(df5)

def distancia_mas_corta(df5):
    inicio=input("Desde donde quieres empezar la ruta?")
    
    ciudades=list(df5.index)

    if inicio not in ciudades:
        print("Esa ciudad no esta en la lista")
        return
    
    sin_visitar=set(ciudades)
    ruta=[inicio]
    sin_visitar.remove(inicio)

    actual=inicio
    distancia=0

    while sin_visitar:
        siguiente=min(
            sin_visitar,
            key=lambda ciudad:df5.loc[actual,ciudad]
        )

        distancia += df5.loc[actual,siguiente]
        ruta.append(siguiente)
        sin_visitar.remove(siguiente)
        actual=siguiente

    return ruta,f" la distancia total es: {distancia} Kms"


distancia_mas_corta(df5)
```
