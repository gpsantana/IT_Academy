```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fondo_res_ss = pd.read_csv(r"C:\Users\GPSantana\Downloads\Fondo de Reserva de las Pensiones.csv",sep=',', encoding="latin1", header=1)

fondo_res_ss=fondo_res_ss.pivot_table(
    columns="Año"
)

fondo_res_ss.columns = ["M€"]

fondo_res_ss = fondo_res_ss.reset_index().rename(columns={"index": "Año"})


fondo_res_ss
```

```python
fondo_res_ss.to_csv(r"C:\Users\GPSantana\Downloads\fondo_res_ss.csv", index=False)
```

```python
plt.figure(figsize=(14,7))
plt.bar(fondo_res_ss["Año"], fondo_res_ss["M€"], color="steelblue")

plt.title("Fondo de Reserva de la Seguridad Social (M€)", fontsize=18)
plt.xlabel("Año", fontsize=14)
plt.ylabel("Millones de euros", fontsize=14)
plt.xticks(rotation=45)

plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()




```
