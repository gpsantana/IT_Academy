```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
evolucion_pens = pd.read_excel(r"C:\Users\GPSantana\Downloads\Evolucion_pensiones1.xls", header=3)

evolucion_pens.head(20)
```

```python
evolucion_pens=evolucion_pens.drop("Unnamed: 1", axis=1)

evolucion_pens = evolucion_pens.drop(evolucion_pens.index[0])

evolucion_pens=evolucion_pens[["PERIODO","TOTAL"]]
```

```python
evolucion_pens.dropna(inplace=True)

evolucion_pens = evolucion_pens.head(10)

evolucion_pens
```

```python
evolucion_pens.to_csv(r"C:\Users\GPSantana\Downloads\evolucion_pensiones.csv", index=False)
```

```python

plt.figure(figsize=(10,5))
sns.lineplot(data=evolucion_pens, x="PERIODO", y="TOTAL", marker="o", linewidth=2)

plt.title("Evolución del número total de pensiones", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Total de pensiones", fontsize=12)
#plt.gca().invert_yaxis()

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
