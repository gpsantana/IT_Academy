4. Fes una petició POST per crear una nova publicació fictícia. Inclou el títol, el cos del missatge i un userId.

Mostra la resposta JSON.
Mostra el codi d'estat

```python
new_post={
    "userID": 666,
    "title": "Olvídate de mi",
    "body": "primer post",
}

response=requests.post('https://jsonplaceholder.typicode.com/todos/',json=new_post)

todos=response.json()

print(todos)

response.status_code
```

5. Fes una petició PATCH per modificar parcialment una publicació existent.

Mostra la resposta JSON.
Mostra el codi d'estat.

```python
update = { "title": "Espero que nunca te olvides de mi"}

response=requests.patch('https://jsonplaceholder.typicode.com/todos/201',json=update)

todos=response.json()

print(todos)

response.status_code
```

6. Fes una petició DELETE sobre una publicació.

Mostra la resposta JSON.
Mostra el codi d'estat.

```python
response=requests.delete('https://jsonplaceholder.typicode.com/todos/201')

response.json()

print(response.text)

response.status_code
```

1. Utilitza l’API d’Open Data BCN per:

Cercar un dataset del teu interès mitjançant package_search i package_show. 

```python

#url= "https://opendata-ajuntament.barcelona.cat/data/api/action/tag_list"
 
url= "https://opendata-ajuntament.barcelona.cat/data/api/action/package_search"

params= {"q":"migrants"}

response=requests.get(url,params=params)

data=response.json()


for dataset in data['result']['results']:
    print(f"- {dataset['title']} (ID: {dataset['id']})")

```

```python
url_show = "https://opendata-ajuntament.barcelona.cat/data/api/action/package_show"
params = {"id": "c427877c-4517-48cb-bcba-5a935bbe0dac"}

response = requests.get(url_show, params=params)
data = response.json()

for resource in data['result']['resources']:
    print(f"Nombre: {resource['name']}")
    print(f"ID: {resource['id']}")
    print(f"Formato: {resource['format']}")

```
